```python
import imgaug.augmenters as iaa
import cv2
import os
```

```python
dataset_path = "D:/Coding/INS_RP/CrossMatch_Sample_DB"
output_path = "D:/Coding/INS_RP/Augmented_CrossMatch_Sample_DB"
```

```python
augmenter = iaa.Sequential([
#     iaa.Fliplr(0.5),  # Flip images horizontally
#     iaa.Flipud(0.5),  # Flip images vertically
    iaa.Affine(rotate=(-10, 10)),  # Rotate images within -10 to 10 degrees
    iaa.GaussianBlur(sigma=(0, 1.0)),  # Apply Gaussian blur with sigma between 0 and 1.0
    iaa.AdditiveGaussianNoise(scale=(0, 0.05 * 255)),  # Add Gaussian noise with a scale of 0 to 0.05
    iaa.Multiply((0.8, 1.2)),  # Multiply pixel values by a random value between 0.8 and 1.2
])
```

```python
files = [file for file in os.listdir(dataset_path)]
for i in range(51):
    set8 = files[:8]
#     print(len(set8))
    files = files[8:]
    last_print = set8.pop(-1)
    image_path = os.path.join(dataset_path, last_print)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    augmented_images = augmenter(images=[image] * 5)  # Generate 5 augmented images for each original image

    for idx, augmented_image in enumerate(augmented_images):
        output_file = f"{last_print[:-4]}_augmented_{idx}.tif"
        output_image_path = os.path.join(output_path, output_file)
        cv2.imwrite(output_image_path, augmented_image)
```
